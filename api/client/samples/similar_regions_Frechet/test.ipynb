```python
%%time
# Full "entire-world" download with default parameters (caution: this takes many hours)
#sim = SimilarRegion(metric_description,
#                    update_mode="reload",
#                    region_info_reload=True
#                    )

# Metric to use
# Here - temperature, rainfall, and soil properties, down-weighted so that
# together the seven of them have roughly the same weight as one of the time series
metric = {
    #'soil_moisture':1.0,
    'rainfall':1.0,
    'land_surface_temperature':1.0,
    'cation_exchange_30cm':1.0/7,
    'ph_h2o_30cm':1.0/7,
    'sand_30cm':1.0/7,
    'silt_30cm':1.0/7,
    'clay_30cm':1.0/7,
    'organic_carbon_content_fine_earth_30cm':1.0/7,
    'soil_water_capacity_100cm':1.0/7
}

# Use cached data only (should take under 5 min for the full world assuming all data is present)
sim = SimilarRegion(metric_description,
                    #search_region=1215, # uncomment to serach over US-only
                    update_mode="no",
                    t_int_per_year=52,
                    metric_instance=metric,
                   drop_mode='any_missing')
```

```python
# Find provinces/districts/countries (depending requested_level) across the world
# (or US if sim object was constructed with search_region=1215)
# similar to California's Napa county
print("Similar to Napa:")
for i in sim.similar_to(136969, number_of_regions=10, requested_level=4):
    print(i)

# Provinces similar to Iowa (with detailed distance report)
print("Similar to Iowa:")
for i in sim.similar_to(13066, number_of_regions=10, requested_level=4,detailed_distance=True):
    print(i)
```
